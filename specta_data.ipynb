```python
# standard curve - SDS Page Analysis

# protein standards for molecular marker ladder, size in (kDa)
markers = [250, 150, 100, 75, 50, 37, 25, 20, 15, 10]

# distance protein taveled on membrane in (cm)
distance = [0.635, 1.8542, 2.54, 4.2418, 5.9944, 8.128, 10.0076, 12.2174, 13.7668, "NaN"]
```

```python
markers_df = pd.DataFrame(markers)
markers_df = markers_df.rename(columns={markers_df.columns[0]: "Molecular Weight (kDa)"})
markers_df
```

```python
distance_df = pd.DataFrame(distance)
distance_df = distance_df.rename(columns={distance_df.columns[0]: "Migration Distance (cm)"})
distance_df
```

```python
# protein standard
df2 = pd.concat([distance_df, markers_df], axis=1)
df2 
```

```python
log_markers_df = (markers_df.apply(np.log10))
log_markers_df
```

```python
log_markers_df = log_markers_df.rename(columns={"Molecular Weight (kDa)" : "Log of Molecular Weight (kDa)"})
log_markers_df
```

```python
# log molecular weight (kDa) vs migration distance (cm) 
df3 = pd.concat([distance_df, markers_df, log_markers_df], axis=1)
df3
```

```python
df3.to_excel (r'C:\Users\chris\mcb-protein_isolation\data\table2.xlsx', index = False, header=True)
```
